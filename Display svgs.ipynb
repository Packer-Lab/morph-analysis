```python
import os
import numpy as np
import pandas as pd
import svgutils.compose as sc
import svgutils.transform as st
import matplotlib.pyplot as plt
%matplotlib notebook
from ams_utilities import *
```

```python
#list .svg files
directory = r"\\163.1.248.177\Root\jswann\Morphologies\Vectorised Reconstruction Images"
datapaths = []
for root, subdirs, files in os.walk(directory):
    for filename in files:
        if filename.endswith(".svg"):
            temppath=os.path.join(root, filename)
            datapaths.append(temppath)
#print(datapaths)

def get_mouseID(datapath):

    mousetemp = datapath.split('_')[0]
    mouse=mousetemp.split('\\')[7]
    celltemp = datapath.split('_')[2]
    cell = celltemp.split(".")[0]
    mouseID = mouse + ': ' + cell

    return mouseID

mouseIDs=[]
for i in datapaths:
    mouseIDs.append(get_mouseID(i))
print(mouseIDs)
```

```python
#Make a dataframe of CellID, .svg datapath
paths=pd.DataFrame(datapaths,index=mouseIDs, columns=['.svg Datapath'])

#Add morph clusters to dataframe
morph_clusters=pd.read_csv(r'\\163.1.248.177\share\Data\Morphologies\morphology_k-means_clusters.csv', index_col=0).drop(['10'],axis=1)
#Rename ccellIDs to match
cellNames=list(morph_clusters.index)
for j in range (1,7):
    cellNames= [i.replace('Cell'+str(j),'Cell0'+str(j)) for i in cellNames]
morph_clusters.index=cellNames

df=pd.concat([paths,morph_clusters], axis=1, sort=True)
df=df.dropna() #Drops cells which have been excluded from analysis
df=df.sort_values(by=['2'])

ephys_data=GS_to_df('https://docs.google.com/spreadsheets/d/1ziOx80em0ZhmMmSjKePYbOq3K6sHcOapfHjy9S4oDbk/edit#gid=1828339698')
triphasic=pd.DataFrame([ephys_data.iloc[0,:],ephys_data.iloc[58,:]])
triphasic.columns=triphasic.iloc[0,:]
triphasic=triphasic.drop([0]).drop(['MouseID: Cell#'],1)
triphasic.index=['Triphasic']
cellNames=list(triphasic.columns)
for j in range (1,7):
    cellNames= [i.replace('Cell'+str(j),'Cell0'+str(j)) for i in cellNames]
triphasic.columns=cellNames
triphasic

df=pd.merge(df,triphasic.T,how='left',left_index=True,right_index=True)
df.to_csv('E:\Jake\SVGs and morph clusters.csv',na_rep='NaN')
df
```
